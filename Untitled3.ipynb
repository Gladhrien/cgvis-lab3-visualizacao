```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
sns.set_theme(style='whitegrid', context='talk')
plt.rcParams['figure.dpi'] = 110

BASE = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2026/2026-04-21'
spend = pd.read_csv(f'{BASE}/health_spending.csv')
fin   = pd.read_csv(f'{BASE}/financing_schemes.csv')
purp  = pd.read_csv(f'{BASE}/spending_purpose.csv')

print('spend:', spend.shape, ' fin:', fin.shape, ' purp:', purp.shape)
print('Years:', spend.year.min(), '→', spend.year.max(),
      ' | countries:', spend.iso3_code.nunique())
spend.head(3)
```

```python
def pivot(df, keycol):
    """Pivot long indicator table into one row per (iso3, year), columns = indicator_code."""
    return (df.pivot_table(index=['iso3_code','country_name','year'],
                           columns='indicator_code', values='value')
              .reset_index())

s = pivot(spend, 'expenditure_type')
f = pivot(fin,   'financing_scheme')
p = pivot(purp,  'spending_purpose')

df = s.merge(f, on=['iso3_code','country_name','year'], how='outer') \
      .merge(p, on=['iso3_code','country_name','year'], how='outer')

# Friendlier display names — official WHO names are long and break chart layouts
NAME_FIXES = {
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'United States of America':                              'United States',
    'Russian Federation':                                    'Russia',
    'Republic of Korea':                                     'South Korea',
    'Democratic People\'s Republic of Korea':                'North Korea',
    'Iran (Islamic Republic of)':                            'Iran',
    'Venezuela (Bolivarian Republic of)':                    'Venezuela',
    'Bolivia (Plurinational State of)':                      'Bolivia',
    'Lao People\'s Democratic Republic':                     'Laos',
    'Syrian Arab Republic':                                  'Syria',
    'Viet Nam':                                              'Vietnam',
    'Czechia':                                               'Czech Republic',
    'Republic of Moldova':                                   'Moldova',
    'United Republic of Tanzania':                           'Tanzania',
    'Democratic Republic of the Congo':                      'DR Congo',
    'Congo':                                                 'Republic of Congo',
    'Brunei Darussalam':                                     'Brunei',
    'occupied Palestinian territory, including east Jerusalem': 'Palestine',
    'Türkiye':                                               'Turkey',
    'Cabo Verde':                                            'Cape Verde',
    'Eswatini':                                              'Eswatini',  # already short
}
df['country_name'] = df['country_name'].replace(NAME_FIXES)
print(f'Renamed {sum(df["country_name"].isin(NAME_FIXES.values()))} rows '
      f'across {df.loc[df["country_name"].isin(NAME_FIXES.values()), "iso3_code"].nunique()} countries')


# Friendly names for the indicators we'll use most
df = df.rename(columns={
    'che_usd2023':   'che_usd',          # total current health expenditure (USD)
    'gghed_usd2023': 'gov_usd',          # government domestic
    'pvtd_usd2023':  'pvt_usd',          # private domestic
    'ext_usd2023':   'ext_usd',          # external (donor) funds
    'gghed_che':     'gov_pct',          # gov % of CHE
    'pvtd_che':      'pvt_pct',
    'ext_che':       'ext_pct',
    'hf1_che':       'gov_scheme_pct',   # compulsory schemes %
    'hf2_che':       'voluntary_pct',    # voluntary %
    'hf3_che':       'oop_pct',          # household out-of-pocket %
})
print(df.shape)
df.head(3)

# Normalize all share columns to 0–100 once, so downstream cells don't multiply
share_cols = [c for c in df.columns
              if c.endswith('_pct') or c.startswith(('hc','hf'))]
for c in share_cols:
    if df[c].max(skipna=True) <= 1.5:
        df[c] = df[c] * 100
print('Normalized to 0–100:', share_cols)

```

```python
# 3. Correlations — split into "size" (USD totals) and "composition" (shares)
LABELS = {
    'che_usd': 'Total CHE (USD)',
    'gov_usd': 'Government (USD)',
    'pvt_usd': 'Private (USD)',
    'ext_usd': 'External (USD)',
    'gov_pct': 'Gov % of CHE',
    'pvt_pct': 'Private % of CHE',
    'ext_pct': 'External % of CHE',
    'oop_pct': 'Out-of-pocket % of CHE',
    'voluntary_pct': 'Voluntary % of CHE',
}
size_cols  = ['che_usd', 'gov_usd', 'pvt_usd', 'ext_usd']
share_cols = ['gov_pct', 'pvt_pct', 'ext_pct', 'oop_pct', 'voluntary_pct']

def tri_heatmap(ax, cols, title):
    have = [c for c in cols if c in df.columns]
    corr = df[have].rename(columns=LABELS).corr()
    mask = np.triu(np.ones_like(corr, dtype=bool), k=1)  # hide upper triangle
    sns.heatmap(corr, mask=mask, annot=True, fmt='.2f',
                cmap='RdBu_r', center=0, vmin=-1, vmax=1,
                square=True, linewidths=.5, linecolor='white',
                cbar_kws={'shrink': .7, 'label': 'Pearson r'},
                annot_kws={'size': 11}, ax=ax)
    ax.set_title(title, pad=12)
    ax.tick_params(axis='x', rotation=30)
    ax.tick_params(axis='y', rotation=0)
    for lbl in ax.get_xticklabels(): lbl.set_ha('right')

fig, (axL, axR) = plt.subplots(1, 2, figsize=(15, 6.5),
                               gridspec_kw={'width_ratios': [1, 1.15]})
tri_heatmap(axL, size_cols,  'Size: USD totals')
tri_heatmap(axR, share_cols, 'Composition: shares of CHE')
plt.tight_layout(); plt.show()
```

```python
# Government share vs out-of-pocket — colored by region
import requests
from itertools import cycle

# --- Fetch ISO3 → region from World Bank metadata ---
meta = requests.get('https://api.worldbank.org/v2/country',
                    params={'format':'json','per_page':400}, timeout=30).json()[1]
region_map = {c['id']: c['region']['value'] for c in meta
              if c.get('region') and c['region']['value'] != 'Aggregates'}

# --- Build latest-year-per-country slice with auto-scaled %s ---
needed = ['gov_pct','oop_pct','che_usd']
latest = (df.dropna(subset=needed)
            .sort_values('year')
            .groupby('iso3_code', as_index=False).tail(1)
            .copy())
def to_pct(s): return s*100 if s.max() <= 1.5 else s
latest['gov_%']  = to_pct(latest['gov_pct'])
latest['oop_%']  = to_pct(latest['oop_pct'])
latest['region'] = latest['iso3_code'].map(region_map).fillna('Unknown')

# --- Auto-derive palette from regions actually in the data ---
present = (latest['region'].value_counts()
                           .drop('Unknown', errors='ignore')
                           .index.tolist())
print(f'regions found ({len(present)}):')
for r in present: print(f'  • {r!r}')

# Colorblind-safe qualitative palette + gray for Unknown
qual = ['#377eb8','#e41a1c','#4daf4a','#984ea3','#ff7f00',
        '#ffd92f','#a65628','#f781bf']
palette = dict(zip(present, cycle(qual)))
palette['Unknown'] = '#bbbbbb'

# --- Plot ---
fig, ax = plt.subplots(figsize=(11, 7))
ax.plot([0,100],[100,0], 'k--', alpha=.25, zorder=1, label='gov + OOP = 100%')

# Plot in stable order: largest region first so smaller ones land on top
for region in present + ['Unknown']:
    sub = latest[latest['region'] == region]
    if sub.empty: continue
    ax.scatter(sub['gov_%'], sub['oop_%'],
               s=60 + 40*(np.log10(sub['che_usd']) - 6),
               color=palette[region],
               alpha=.78, edgecolor='k', linewidth=.4, zorder=3,
               label=region)

# Label biggest spenders + worst-OOP
for _, r in latest.nlargest(8, 'che_usd').iterrows():
    ax.annotate(r['country_name'][:18], (r['gov_%'], r['oop_%']),
                fontsize=9, alpha=.85, xytext=(5,3), textcoords='offset points')
for _, r in latest.nlargest(3, 'oop_%').iterrows():
    ax.annotate(r['country_name'][:18], (r['gov_%'], r['oop_%']),
                fontsize=9, color='crimson', alpha=.9,
                xytext=(5,3), textcoords='offset points')

ax.set(title='Government share vs out-of-pocket share\n(latest year per country, colored by region)',
       xlabel='Government schemes (% of CHE)', ylabel='Out-of-pocket (% of CHE)',
       xlim=(-3,103), ylim=(-3,103))
ax.text(0.02, 0.97, 'Bubble size = total CHE',
        transform=ax.transAxes, fontsize=10, alpha=.7, va='top')
ax.legend(loc='upper right', fontsize=8.5, framealpha=.92)
plt.tight_layout(); plt.show()
```

```python
# Top 12 spenders, with the "private" bar split into voluntary insurance vs OOP
top = (df.query('year == 2022').dropna(subset=['che_usd'])
         .nlargest(12, 'che_usd')
         .sort_values('che_usd'))

# Compute USD per bucket from the % shares (auto-scale 0–1 vs 0–100)
def to_pct(s): return s*100 if s.max(skipna=True) <= 1.5 else s
for col in ['gov_pct','voluntary_pct','oop_pct','ext_pct']:
    top[col + '_n'] = to_pct(top[col]) / 100.0
top['gov_$']  = top['che_usd'] * top['gov_pct_n']
top['vol_$']  = top['che_usd'] * top['voluntary_pct_n'].fillna(0)
top['oop_$']  = top['che_usd'] * top['oop_pct_n'].fillna(0)
top['ext_$']  = top['che_usd'] * top['ext_pct_n'].fillna(0)
top['other_$'] = (top['che_usd']
                  - top[['gov_$','vol_$','oop_$','ext_$']].sum(axis=1)).clip(lower=0)

fig, ax = plt.subplots(figsize=(12, 6.5))
left = np.zeros(len(top))
buckets = [
    ('gov_$',   'Government',                     '#2b6cb0'),
    ('vol_$',   'Voluntary (private insurance)',  '#dd6b20'),
    ('oop_$',   'Out-of-pocket',                  '#c53030'),
    ('ext_$',   'External (donors)',              '#38a169'),
    ('other_$', 'Other / unspecified',            '#888888'),
]
for col, lab, color in buckets:
    vals = (top[col].fillna(0) / 1e9).values
    ax.barh(top['country_name'], vals, left=left, label=lab, color=color, alpha=.92)
    left += vals

for i, t in enumerate(top['che_usd'].values):
    ax.text(t/1e9 + 35, i, f' ${t/1e9:,.0f}B', va='center', fontsize=10, alpha=.7)

ax.set(title='Top 12 countries by current health expenditure, 2022',
       xlabel='CHE (billion 2023 USD)', ylabel='')
ax.legend(loc='lower right', fontsize=9)
plt.tight_layout(); plt.show()
```

```python
# 8. Spending-purpose composition for the biggest spenders
purpose_cols = {
    'hc1_che': 'Curative care',
    'hc2_che': 'Rehabilitative',
    'hc3_che': 'Long-term care',
    'hc4_che': 'Ancillary services',
    'hc5_che': 'Medical goods',
    'hc6_che': 'Preventive care',
    'hc7_che': 'Governance/admin',
    'hc9_che': 'Other (n.e.c.)',
}
have = [c for c in purpose_cols if c in df.columns]

# Latest year per country with at least 2 hc* values populated
purpose_2022 = df.query('year == 2022').dropna(subset=have, thresh=2)

big = (purpose_2022.merge(df.query('year == 2022')[['iso3_code','che_usd']],
                          on='iso3_code', suffixes=('','_x'))
                   .nlargest(10, 'che_usd')['iso3_code'].tolist())

sub = purpose_2022[purpose_2022['iso3_code'].isin(big)].copy()
def to_pct(s): return s*100 if s.max(skipna=True) <= 1.5 else s
for c in have:
    sub[c] = to_pct(sub[c])

sub = sub.set_index('country_name')[have].rename(columns=purpose_cols)
sub = sub.loc[sub.sum(axis=1).sort_values(ascending=False).index]

# Also list which big spenders were excluded for transparency
excluded = (df.query('year == 2022').dropna(subset=['che_usd'])
              .nlargest(15, 'che_usd')[['country_name','iso3_code']]
              .pipe(lambda d: d[~d['iso3_code'].isin(big)]))
print('Top-15 spenders with no purpose breakdown in 2022:')
for _, r in excluded.iterrows(): print(f'  • {r["country_name"]}')

fig, ax = plt.subplots(figsize=(11, 6))
sub.plot(kind='barh', stacked=True, colormap='tab20', ax=ax, width=.8)
ax.set(title='What top spenders spend on, 2022 (% of CHE)\n— excludes countries that don\'t report this breakdown',
       xlabel='% of current health expenditure', ylabel='', xlim=(0, 100))
ax.legend(title='Spending purpose', bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=9)
plt.tight_layout(); plt.show()
```

```python
delta = (df.dropna(subset=['gov_pct'])
           .groupby('iso3_code')
           .apply(lambda g: pd.Series({
               'country':  g['country_name'].iloc[0],
               'start':    g.loc[g['year'].idxmin(), 'gov_pct'],
               'end':      g.loc[g['year'].idxmax(), 'gov_pct'],
               'years':    f"{g['year'].min()}–{g['year'].max()}",
           })))
delta['shift'] = (delta['end'] - delta['start']) * 100

movers = pd.concat([delta.nlargest(8,'shift'), delta.nsmallest(8,'shift')]).sort_values('shift')
fig, ax = plt.subplots(figsize=(10,7))
colors = ['#2b6cb0' if s > 0 else '#c53030' for s in movers['shift']]
ax.barh(movers['country'], movers['shift'], color=colors, alpha=.85)
ax.axvline(0, color='k', lw=.8)
for i,(c,s) in enumerate(zip(movers['country'], movers['shift'])):
    ax.text(s + (1 if s>0 else -1), i, f'{s:+.1f} pp', va='center',
            ha='left' if s>0 else 'right', fontsize=9, alpha=.8)
ax.set(title='Biggest shifts in government share of health spending\n(percentage points, first vs last available year)',
       xlabel='Δ Government % of CHE (pp)', ylabel='')
plt.tight_layout(); plt.show()
```

```python
# Preventive vs curative spending vs GDP per capita — one panel, raw scatter
import requests
import matplotlib.ticker as mticker

url = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.CD'
r = requests.get(url, params={'date': '2022', 'format': 'json', 'per_page': 400},
                 timeout=30).json()
gdp = pd.DataFrame([{'iso3_code': d['countryiso3code'], 'gdp_pc': d['value']}
                    for d in r[1] if d.get('countryiso3code') and d.get('value')])

m = (df.query('year == 2022')
       .merge(gdp, on='iso3_code', how='inner')
       .dropna(subset=['hc1_che', 'hc6_che', 'gdp_pc']))
def to_pct(s): return s*100 if s.max() <= 1.5 else s
m['curative']   = to_pct(m['hc1_che'])
m['preventive'] = to_pct(m['hc6_che'])
print(f'Countries: {len(m)}')

CURATIVE   = '#dd6b20'
PREVENTIVE = '#38a169'

fig, ax = plt.subplots(figsize=(12, 7))

ax.scatter(m['gdp_pc'], m['curative'],   s=60, color=CURATIVE,   alpha=.65,
           edgecolor='k', linewidth=.3, label='Curative care')
ax.scatter(m['gdp_pc'], m['preventive'], s=60, color=PREVENTIVE, alpha=.65,
           edgecolor='k', linewidth=.3, label='Preventive care')

# Log GDP axis with $ formatted ticks
ax.set_xscale('log')
ax.xaxis.set_major_locator(mticker.LogLocator(base=10, numticks=6))
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))

# Spearman ρ for both, in the title
rho_c = m[['gdp_pc','curative']].corr('spearman').iloc[0,1]
rho_p = m[['gdp_pc','preventive']].corr('spearman').iloc[0,1]
ax.set(title=f'Curative vs preventive care by GDP per capita, 2022\n'
             f'(Spearman ρ: curative {rho_c:+.2f}, preventive {rho_p:+.2f}, n={len(m)})',
       xlabel='GDP per capita, USD (log scale)',
       ylabel='% of current health expenditure')
ax.legend(loc='upper right')
plt.tight_layout(); plt.show()
```

```python
# OOP: low as a share, high in absolute dollars
import requests

# 1. Fetch population from World Bank for 2022
pop_r = requests.get('https://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL',
                     params={'date':'2022','format':'json','per_page':400}, timeout=30).json()
pop = pd.DataFrame([{'iso3_code': d['countryiso3code'], 'pop': d['value']}
                    for d in pop_r[1] if d.get('countryiso3code') and d.get('value')])

# 2. OOP USD per capita: hf3_usd2023 is OOP in constant USD
m = (df.query('year == 2022')
       .merge(pop, on='iso3_code', how='inner')
       .dropna(subset=['hf3_usd2023','oop_pct','pop']))
m['oop_pc'] = m['hf3_usd2023'] / m['pop']
def to_pct(s): return s*100 if s.max() <= 1.5 else s
m['oop_share'] = to_pct(m['oop_pct'])

# 3. Two-panel: top by share, top by per-capita dollars
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

# Left: top-15 by share
left = m.nlargest(15,'oop_share').sort_values('oop_share')
us_l = left['iso3_code']=='USA'
axes[0].barh(left['country_name'], left['oop_share'],
             color=['#c53030' if u else '#888' for u in us_l], alpha=.85)
axes[0].set(title='Top 15 by OOP share of CHE\n(what % comes directly from households)',
            xlabel='OOP (% of CHE)')

# Right: top-15 by per-capita USD, with US highlighted even if not top-15
right = m.nlargest(15,'oop_pc')
if 'USA' not in right['iso3_code'].values:
    right = pd.concat([right, m[m['iso3_code']=='USA']])
right = right.sort_values('oop_pc')
us_r = right['iso3_code']=='USA'
axes[1].barh(right['country_name'], right['oop_pc'],
             color=['#c53030' if u else '#888' for u in us_r], alpha=.85)
for i,v in enumerate(right['oop_pc']):
    axes[1].text(v+15, i, f'${v:,.0f}', va='center', fontsize=9, alpha=.75)
axes[1].set(title='Top 15 by OOP $ per capita\n(how many actual dollars households spend)',
            xlabel='OOP USD per capita, 2022')

us_row = m[m['iso3_code']=='USA'].iloc[0]
fig.suptitle(f'USA: OOP is {us_row["oop_share"]:.1f}% of CHE  '
             f'but ${us_row["oop_pc"]:,.0f}/person/year', fontsize=13, y=1.02)
plt.tight_layout(); plt.show()
```

```python
# Brazil / SUS in the OOP picture — trajectory + peer comparison
import requests
import matplotlib.ticker as mticker

# --- Population for per-capita math ---
pop_r = requests.get('https://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL',
                     params={'date': '2000:2023', 'format':'json', 'per_page': 20000},
                     timeout=30).json()
pop = pd.DataFrame([{'iso3_code': d['countryiso3code'],
                     'year': int(d['date']),
                     'pop': d['value']}
                    for d in pop_r[1] if d.get('countryiso3code') and d.get('value')])

# --- Brazil time series ---
def to_pct(s): return s*100 if s.max(skipna=True) <= 1.5 else s
br = (df.query("iso3_code == 'BRA'")
        .merge(pop, on=['iso3_code','year'], how='left')
        .sort_values('year').copy())
br['gov_%']        = to_pct(br['gov_pct'])
br['oop_%']        = to_pct(br['oop_pct'])
br['voluntary_%']  = to_pct(br['voluntary_pct'])
br['oop_pc_usd']   = br['hf3_usd2023'] / br['pop']

# --- Peer set: same year (latest available with all 3 fields) ---
peers = ['BRA','USA','GBR','NOR','MEX','ARG','CHL','IND','CRI','CUB']
latest = (df.dropna(subset=['oop_pct','hf3_usd2023'])
            .sort_values('year')
            .groupby('iso3_code', as_index=False).tail(1))
latest = latest[latest['iso3_code'].isin(peers)].merge(pop, on=['iso3_code','year'], how='left').copy()
latest['oop_%']      = to_pct(latest['oop_pct'])
latest['oop_pc_usd'] = latest['hf3_usd2023'] / latest['pop']
latest = latest.sort_values('oop_%')

# --- Plot: 2 panels ---
fig, (axL, axR) = plt.subplots(1, 2, figsize=(15, 6.5))

# LEFT: Brazil's financing-mix shares over time
axL.plot(br['year'], br['gov_%'],       lw=3, color='#2b6cb0', label='Government schemes')
axL.plot(br['year'], br['voluntary_%'], lw=3, color='#dd6b20', label='Voluntary (private insurance)')
axL.plot(br['year'], br['oop_%'],       lw=3, color='#c53030', label='Out-of-pocket')
axL.fill_between(br['year'], 0, br['oop_%'], color='#c53030', alpha=.10)

# Annotate latest year
last = br.dropna(subset=['gov_%','oop_%','voluntary_%']).iloc[-1]
for col, color in [('gov_%','#2b6cb0'), ('voluntary_%','#dd6b20'), ('oop_%','#c53030')]:
    axL.annotate(f'{last[col]:.0f}%', (last['year'], last[col]),
                 color=color, fontsize=11, fontweight='bold',
                 xytext=(6, 0), textcoords='offset points', va='center')

axL.set(title="Brazil's health-spending mix, 2000–2023\n(SUS = government schemes; ANS-regulated private = voluntary)\n",
        xlabel='Year', ylabel='% of current health expenditure', ylim=(0, 70))
axL.yaxis.set_major_formatter(mticker.PercentFormatter(decimals=0))
axL.legend(loc='center right', fontsize=10)

# RIGHT: peer comparison — OOP share AND OOP $ per capita
y = np.arange(len(latest))
axR.barh(y - 0.2, latest['oop_%'],      height=0.4,
         color=['#c53030' if c=='BRA' else '#888' for c in latest['iso3_code']],
         alpha=.85, label='OOP % of CHE')
axR.set_xlabel('OOP (% of CHE)', color='#444')
axR.set_yticks(y); axR.set_yticklabels(latest['country_name'])
for i, v in enumerate(latest['oop_%']):
    axR.text(v + 0.6, i - 0.2, f'{v:.0f}%', va='center', fontsize=9, alpha=.7)

# Twin axis for $ per capita
axR2 = axR.twiny()
axR2.barh(y + 0.2, latest['oop_pc_usd'], height=0.4,
          color=['#c53030' if c=='BRA' else '#bbb' for c in latest['iso3_code']],
          alpha=.55, hatch='//', edgecolor='white', label='OOP $/person/yr')
axR2.set_xlabel('OOP USD per capita', color='#444')
for i, v in enumerate(latest['oop_pc_usd']):
    axR2.text(v + max(latest['oop_pc_usd'])*0.01, i + 0.2,
              f'${v:,.0f}', va='center', fontsize=9, alpha=.7)

axR.set_title("Brazil vs peers — OOP share AND OOP dollars\n(top = $/person/yr, bottom = % of oop)")

# Brazil pull-out annotation
br_row = latest[latest['iso3_code']=='BRA'].iloc[0]
fig.suptitle(f"Brazil's SUS: tax-funded universal coverage, but ~25% of Brazilians also carry private plans  |  "
             f"OOP = {br_row['oop_%']:.0f}% of CHE, ${br_row['oop_pc_usd']:.0f}/person/yr",
             fontsize=12, y=1.02)
plt.tight_layout(); plt.show()
```
